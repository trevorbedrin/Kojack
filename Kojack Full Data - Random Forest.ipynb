```python
headers = list(filtered_df.columns.values)

headers.remove('Month')
headers.remove('Weekday')

final_df = filtered_df[headers]

headers.remove('Users')
print len(headers)
```

```python
X = preprocessing.scale(final_df[headers])
#X = final_df[headers]
y = final_df['Users']
```

```python
X_training, X_testing, y_training, y_testing = train_test_split(
     X, y, test_size=0.33)
```

```python
model = ensemble.RandomForestRegressor()
```

```python
model.fit(X, y)    
```

```python
# calculate R^2
print "R-Squared: %.3f" % r2_score(y_testing, model.predict(X_testing))

# Plot predictions versus actuals
plt.figure(figsize=[15,5])

# plot the predictions 
plt.plot(y_testing, model.predict(X_testing), 'go')
# plot the 1:1 actual outcome line
plt.plot(y_testing, y_testing, 'b')
# axis labels
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
